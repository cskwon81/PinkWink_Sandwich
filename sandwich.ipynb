```python
#매장 상세정보찾기

shop_url = []
shop_add = []
shop_tel = []
shop_menu = []
shop_price = []

for n in df.index[:10]:
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    #매장 URL
    #매장 홈페이지라는 단어를 조회해서 있으면 추가, 없으면 홈페이지 없음으로 표시
    if soup_tmp.find('h4',string='매장 홈페이지'):
        shop_url.append(soup_tmp.find_all('a','link')[2].get_text())
    else:
        shop_url.append('홈페이지 없음')
    
    #매장 주소
    shop_add.append(soup_tmp.find('a','txt_adr').get_text())
    
    #전화번호
    shop_tel.append(soup_tmp.find('div','p_tel').find('p').get_text())
    
    #메뉴
    shop_menu.append(soup_tmp.find('h4',string='대표메뉴').find_next('span','tit').get_text())
    
    #가격
    shop_price.append(soup_tmp.find('h4',string='대표메뉴').find_next('label').get_text())    
```

```python
# 상태 진행바를 만들어주는 tqdm 모듈
# study36 환경 -> pip install tqdm

from tqdm import tqdm_notebook

#매장 상세정보찾기

shop_url = []
shop_add = []
shop_tel = []
shop_menu = []
shop_price = []

# tqdm으로 상태 진행바 표시
for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    #매장 URL
    #매장 홈페이지라는 단어를 조회해서 있으면 추가, 없으면 홈페이지 없음으로 표시
    if soup_tmp.find('h4',string='매장 홈페이지'):
        shop_url.append(soup_tmp.find_all('a','link')[2].get_text())
    else:
        shop_url.append('홈페이지 없음')
    
    #매장 주소
    shop_add.append(soup_tmp.find('a','txt_adr').get_text())
    
    #전화번호
    shop_tel.append(soup_tmp.find('div','p_tel').find('p').get_text())
    
    #메뉴
    shop_menu.append(soup_tmp.find('h4',string='대표메뉴').find_next('span','tit').get_text())
    
    #가격
    shop_price.append(soup_tmp.find('h4',string='대표메뉴').find_next('label').get_text())  
```

```python
#카카오맵 open api 이용해서 주소를 gps 좌표로 변환

import requests
```

```python
add = df['address'][1]
url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + add
rest_api_key = 'e136db6bd8eabd311ae2adbad82f7e3a'
head = {'Authorization': 'KakaoAK ' + rest_api_key}

r = requests.get(url, headers=head)


```

```python
def getLatLng(address):
    result = ''
    
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    rest_api_key = 'e136db6bd8eabd311ae2adbad82f7e3a'
    header = {'Authorization': 'KakaoAK ' + rest_api_key}
    
    r = requests.get(url, headers=header)
    if r.status_code == 200:
        result_address = r.json()['documents'][0]['address']
        
        result = result_address['y'], result_address['x']
    else:
        result = 'Error[' + str(r.status_code) + ']'
            
    return result

getLatLng(df['address'][1])
    
```

```python
# 10개 매장의 gps 좌표 계산
lat = []
lng = []

for n in tqdm_notebook(df.index):
    lat.append(getLatLng(df['address'][n])[0])
    lng.append(getLatLng(df['address'][n])[1])

```

```python
# 데이터프레임에 추가
df['lat'] = lat
df['lng'] = lng
df.info()
```

```python
df['lat'] = pd.to_numeric(df['lat'])
df.info()
```

```python
df['lng'] = pd.to_numeric(df['lng'])
df.info()
```

```python
# folium으로 지도에 표시 (10개 매장의 평균 위치로 불러오기)
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                    zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()],
             popup='center').add_to(mapping)
mapping
```

```python
# 10개 매장을 지도에 표시하고 팝업 설정
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                    zoom_start=11)

for n in tqdm_notebook(df.index):
    folium.Marker([df['lat'][n], df['lng'][n]], popup=df['Cafe'][n]).add_to(mapping)
    
mapping
```
